```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv('marketing_customer_analysis.csv')
data
```

```python
data.shape
```

```python
pd.DataFrame(data)

data.describe()
```

```python
data.columns = data.columns.str.lower()
data.columns
```

```python
plt.hist(data['response'])
plt.show()
```

```python

data['response_rate'] = np.where(data['response']=='Yes', 1, 0)
data
```

```python
#Show a plot of the response rate by the sales channel.
data.groupby('sales channel')['response_rate'].mean().plot.bar()
plt.show()
```

```python
#Show a plot of the response rate by the total claim amount.
data1 = data[(data['response']=='Yes') | (data['response'] == 'No')]
sns.boxplot(x='response', y='total claim amount', data=data1)
plt.show()
```

```python
sns.boxplot(x='response', y='income', data=data1)
plt.show()
```
